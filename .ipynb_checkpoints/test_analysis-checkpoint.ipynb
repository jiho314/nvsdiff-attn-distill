```python
cond_num =[1,2]
value ="lpips"
# --
df = data[(data["cond_num"].isin(cond_num)) & (~data['train_step'].isin([ ]))  ] # & (data["view_range"] == 10) #& (data["train_step"] != 60000)  120000 80000 40000 20000 10000
# df = df[(data["cond_num"].isin([cond_num])
# df = data
# df = data
# --- Group by run_name + train_step, average psnr ---
df_avg_psnr = df.groupby(["run_name", "train_step"])["psnr"].mean().reset_index()
df_avg_ssim = df.groupby(["run_name", "train_step"])["ssim"].mean().reset_index()
df_avg_lpips = df.groupby(["run_name", "train_step"])["lpips"].mean().reset_index()
df_avg_psnr_masked = df.groupby(["run_name", "train_step"])["psnr_masked"].mean().reset_index()

plot_df = eval(f"df_avg_{value}")
plt.figure(figsize=(8,6))
for run in plot_df["run_name"].unique():
    subset = plot_df[plot_df["run_name"] == run]
    plt.plot(subset["train_step"], subset[value], marker="o", label=run)


plt.xlabel("Train Step")
plt.ylabel(f"{value} (avg)")
plt.title(f"{value} (cond_num ={cond_num})")#(view_range=(6/10)(10/14))
plt.legend()
plt.grid(True)
plt.show()
```
